```python
from kraken import binarization, pageseg, rpred
from kraken.lib import models
from PIL import Image
import pandas as pd
import os

def load_model(model_path):
    print("Loading model...")
    return models.load_any(model_path)

def extract_text(image_path, model):
    try:
        image = Image.open(image_path)
        image = binarization.nlbin(image)
        seg = pageseg.segment(image)
        pred = rpred.rpred(model, image, seg)
        text = "\n".join([line.prediction for line in pred])
        return text
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return ""

if __name__ == "__main__":
    model_path = "/content/drive/MyDrive/ArabicManuscriptsDigitization.mlmodel"
    images_folder = "/content/drive/MyDrive/testing images/test"
    csv_path = "/content/drive/MyDrive/testing images/test_df.csv"

    df = pd.read_csv(csv_path)
    model = load_model(model_path)

    results = []

    for _, row in df.iterrows():
        image_name = row['image']
        image_path = os.path.join(images_folder, image_name)

        predicted_text = extract_text(image_path, model)
        results.append({'image': image_name, 'predicted_text': predicted_text})

    # Save results to CSV
    output_df = pd.DataFrame(results)
    output_df.to_csv("/content/drive/MyDrive/Arabic-Manuscripts-predictio.csv", index=False)
    print("Saved predictions to Arabic-Manuscripts-Digitization-Prediction.csv")
```
